```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shubham2703/five-crop-diseases-dataset")

print("Path to dataset files:", path)
```

```python
import os, cv2, matplotlib.pyplot as plt, glob

# Show the first few subfolders (classes)
classes = os.listdir(path)
print("Found classes:", classes)

# Pick one class (e.g., first one)
sample_class = classes[0]
print("Showing samples from:", sample_class)

# Find some image files inside that class
image_files = glob.glob(os.path.join(path, sample_class, "*.jpg"))
print(f"Found {len(image_files)} images in {sample_class}")

# Display a few sample images
plt.figure(figsize=(12, 4))
for i, img_path in enumerate(image_files[:3]):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(1, 3, i+1)
    plt.imshow(img)
    plt.title(sample_class)
    plt.axis("off")
plt.show()
```
